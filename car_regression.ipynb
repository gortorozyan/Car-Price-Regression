```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
```

```python
car_data = pd.read_csv('car_data.csv')
car_data
```

```python
car_data[car_data.Fuel_Type.unique()] = car_data[car_data.Seller_Type.unique()] = car_data[car_data.Transmission.unique()] = 0 
for i in car_data.Fuel_Type:
    car_data[i] = (car_data.Fuel_Type == i).astype('int8') * 1
for i in car_data.Seller_Type:
    car_data[i] = (car_data.Seller_Type == i).astype('int8') * 1
for i in car_data.Transmission:
    car_data[i] = (car_data.Transmission == i).astype('int8') * 1
```

```python
car_data.drop(['Seller_Type', 'Transmission', 'Present_Price', 'Owner', 'Car_Name', 'Fuel_Type'], axis=1, inplace=True)
```

```python
car_data
```

```python
x = car_data.drop('Selling_Price',axis=1)
y = car_data.Selling_Price
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=3)
```

```python
model = RandomForestRegressor(n_estimators=100, random_state=3)
model.fit(x_train, y_train)
```

```python
predictions = model.predict(x_test)
```
